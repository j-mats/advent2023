```python
# numeric word library
numeric_dict = {}
# singles
units = [
"zero", "one", "two", "three", "four", "five", "six", "seven", "eight",
"nine", "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen",
"sixteen", "seventeen", "eighteen", "nineteen",
]

# tens
tens = ["", "", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]

for idx, word in enumerate(units): numeric_dict[word] = (idx)
for idx, word in enumerate(tens): numeric_dict[word] = (idx * 10)

numeric_dict.pop('')
```

```python
numeric_dict
```

```python
import re

def parse_int(text, dictionary=numeric_dict):
    intermediary = []
    result: list = []
    # loop while checking text for matches to dict keys
    for word in dictionary:

        if word in text:
            indx = [ix for ix in range(len(text)) if text.find(word, ix) == ix]
            # print(word, indx)
            number = dictionary[word]
            # print(word, number, indx)
            current = str(number)
            if len(indx) > 1:
                # print('multiple')
                for idx in indx:
                    if len(current) > 1:
                        intermediary.extend([(idx + current.index(i), i) for i in current])
                    else:
                        intermediary.append((idx, current))
                    # print(idx, current)
            else:
                # print('single')
                if len(current) > 1:
                    intermediary.extend([(indx[0] + current.index(i), i) for i in current])
                else:
                    intermediary.append((indx[0], current))
        
    intermediary.sort()
    # print(intermediary)
    result.extend([num for _, num in intermediary])

    # return the result plus the current
    return result

def parse_alnum(text):
    result = []
    if text.isalnum():
        # print(text)
        r = re.split(r'(\d)', text)
        r = [s for s in r if len(s) > 0]
        # print(r)
        for elem in r:
            if elem.isalpha():
                result.extend(parse_int(elem))
            else:
                result.append(elem)
    else:
        result.extend(parse_int(text))
        # print(result)
    return result
```

```python
test_entry = '''Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue
Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red
Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red
Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green
'''
```

```python
listified = test_entry.splitlines()
```

```python
l = []
for x in listified[0].split(': ')[1].split('; '):
    for s in x.split(', '):
        l.append(tuple(s.split(' ')))

d = {}
for n, k in l:
    if (k not in d) or (d[k] < int(n)):
        d.update({k: int(n)})
    else: continue
print(l)
d
```

```python
def parse_game_data(entry_input):
    game_dict = {}
    for entry in entry_input.splitlines():
        game, trials = entry.split(': ')
        # Create a list of tuples (freq, colour)
        t = [tuple(trial.split(' ')) for elem in trials.split('; ') for trial in elem.split(', ')]
        # Use a dictionary comprehension to create a dictionary with the maximum frequency for each colour
        trial_dict = {colour: max(int(freq) for freq, clr in t if clr == colour) for _, colour in t}
        game_dict.update({game: trial_dict})
    return game_dict    

# Save output to variable
game_requirement_dict = parse_game_data(test_entry)
```

```python
def check_against_criteria(game_dict: dict, criteria: dict, debug: bool = False):
    accepted_game_numbers = []
    for game in game_dict:
        colours_within_limits = 0
        trial_check = game_dict[game]
        for colour in criteria:
            if criteria[colour] >= trial_check[colour]:
                colours_within_limits += 1
            else: continue
        if colours_within_limits == 3:
            accepted_game_numbers.append(int(game.replace('Game ', '')))

    print(f'Number of accepted games: {len(accepted_game_numbers)}')
    print(f'Accepted games: {list(map(lambda x: 'Game ' + str(x), accepted_game_numbers))}')
    print(f'Sum of IDs of the accepted games: {sum(accepted_game_numbers)}')
    if debug:
        return accepted_game_numbers, sum(accepted_game_numbers)

criterion = {
    'blue': 14,
    'green': 13,
    'red': 12
}
check_against_criteria(parse_game_data(test_entry), criterion)
```

```python
def get_minimum_cubes(game_dict: dict, debug: bool = False):
    minimum_power = {}
    for game in game_dict:
        prod = 1
        for colour in game_dict[game]:
            # print(colour)
            prod = prod * game_dict[game][colour]
        minimum_power.update({f'{game} Power': prod})

    if debug:
        print(minimum_power)
    print(f'Sum of the power of the games: {sum(minimum_power.values())}')

get_minimum_cubes(game_requirement_dict, True)
```
